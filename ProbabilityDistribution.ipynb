# Probability Distribution

## Basic Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.special import comb, factorial
%matplotlib inline
```

## Mean and Variance

```python
def mean(n, p):
    return sum(n * p) * (n[1] - n[0])

def var(n, p):
    mu = mean(n, p)
    return sum((n - mu)**2 * p) * (n[1] - n[0])
```

# Binomial 

```python
def binomial(k, n, p):
    return comb(n, k) * p**k * (1-p)**(n-k)
```

```python
n = 50
k = np.arange(n)
b85 = binomial(k, n, 0.85)
b50 = binomial(k, n, 0.50)
b30 = binomial(k, n, 0.30)
plt.plot(k, b30, 'o', label = "$p = 0.30$")
plt.plot(k, b50, 'o', label = "$p = 0.50$")
plt.plot(k, b85, 'o', label = "$p = 0.85$")
plt.legend()
plt.show()
```

```python
plt.plot(k, np.cumsum(b30), 'o')
plt.plot(k, np.cumsum(b50), 'o')
plt.plot(k, np.cumsum(b85), 'o')
plt.show()
```

## Poisson

```python
def poisson(k, l):
    return (np.exp(-l) * l**k) / factorial(k)
```

```python
k = np.arange(50)
p025 = poisson(k, 02.5)
p050 = poisson(k, 05.0)
p200 = poisson(k, 20.0)
plt.plot(k,p025, 'o', label='$\lambda =  2.5$')
plt.plot(k,p050, 'o', label='$\lambda =  5.0$')
plt.plot(k,p200, 'o', label='$\lambda = 20.0$')
plt.legend()
plt.show()
```

```python
plt.plot(k, np.cumsum(p025), 'o')
plt.plot(k, np.cumsum(p050), 'o')
plt.plot(k, np.cumsum(p200), 'o')
plt.show()
```

## Exponential 

```python
def exponential(x, l):
    return l * np.exp(-l * x)
```

```python
x = np.linspace(0, 20, 5000)
e05 = exponential(x, 0.5)
e10 = exponential(x, 1.0)
e15 = exponential(x, 1.5)
plt.plot(x, e05, label="$\lambda = 0.5$")
plt.plot(x, e10, label="$\lambda = 1.0$")
plt.plot(x, e15, label="$\lambda = 1.5$")
plt.legend()
plt.show()
```

```python
plt.plot(x, np.cumsum(e05)*(x[1] - x[0]), label="$\lambda = 0.5$")
plt.plot(x, np.cumsum(e10)*(x[1] - x[0]), label="$\lambda = 1.0$")
plt.plot(x, np.cumsum(e15)*(x[1] - x[0]), label="$\lambda = 1.5$")
plt.legend()
plt.show()
```

## Normal

```python
def normal(x, m, s):
    return np.exp(-0.5*((x - m)/s)**2)/(s * np.sqrt(2*np.pi))
```

```python
x = np.linspace(-10, 10, 10000)
n_0_05 = normal(x, 0, 0.5)
n_0_10 = normal(x, 0, 1.0)
n_0_20 = normal(x, 0, 2.0)
n_5_09 = normal(x, 5, 0.9)
n_1_15 = normal(x, 1, 1.5)
n_2_07 = normal(x, 2, 0.7)
plt.plot(x, n_0_05, label = "$\mu = 0, \sigma = 0.5$")
plt.plot(x, n_0_10, label = "$\mu = 0, \sigma = 1.0$")
plt.plot(x, n_0_20, label = "$\mu = 0, \sigma = 2.0$")
plt.legend()
plt.show()


plt.plot(x, n_5_09, label = "$\mu = 5, \sigma = 0.9$")
plt.plot(x, n_1_15, label = "$\mu = 1, \sigma = 1.5$")
plt.plot(x, n_2_07, label = "$\mu = 2, \sigma = 0.7$")
plt.legend()
plt.show()
```

```python
plt.plot(x, np.cumsum(n_0_05)*(x[1] - x[0]), label = "$\mu = 0, \sigma = 0.5$")
plt.plot(x, np.cumsum(n_0_10)*(x[1] - x[0]), label = "$\mu = 0, \sigma = 1.0$")
plt.plot(x, np.cumsum(n_0_20)*(x[1] - x[0]), label = "$\mu = 0, \sigma = 2.0$")
plt.legend()
plt.show()


plt.plot(x, np.cumsum(n_5_09)*(x[1] - x[0]), label = "$\mu = 5, \sigma = 0.9$")
plt.plot(x, np.cumsum(n_1_15)*(x[1] - x[0]), label = "$\mu = 1, \sigma = 1.5$")
plt.plot(x, np.cumsum(n_2_07)*(x[1] - x[0]), label = "$\mu = 2, \sigma = 0.7$")
plt.legend()
plt.show()
```
