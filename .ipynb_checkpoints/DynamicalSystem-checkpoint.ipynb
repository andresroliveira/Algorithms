# Sistemas Dinâmicos

## Bibliotecas Básicas 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

Queremos resolver problemas da forma 
$$ \dot{x} = f(x) $$ 
com a condição inicial de que $x(0) = x_0$ e $x \in \mathbb{R}^n$.

<!-- Faremos aqui problemas com $n=1$ e $n=2$. -->

## Definindo a Função

Queremos resolver:
$$ \begin{cases} 
    \dot{x} &= -y \\
    \dot{y} &= x
\end{cases}$$
com a condição de que $$ (x_0, y_0) = (1, 0)$$

```python
x_0 = 1
y_0 = 0

def f(x, y):
    return -y, x
```

### Método de Euler

```python
n = 2*314
dt = 0.01

x = np.zeros(n)
y = np.zeros(n)

x[0] = x_0
y[0] = y_0

for i in range(1,n):
    dx, dy = f(x[i-1], y[i-1])
    x[i] = x[i-1] + dx*dt
    y[i] = y[i-1] + dy*dt

plt.plot(x,y)
plt.show()
```

## Método de Heun

```python
n = 2*314
dt = 0.01

x = np.zeros(n)
y = np.zeros(n)

x[0] = x_0
y[0] = y_0

for i in range(1,n):
    dx_1, dy_1 = f(x[i-1], y[i-1])
    dx_2, dy_2 = f(x[i-1] + dt*dx_1, y[i-1]+dt*dy_1)
    
    x[i] = x[i-1] + dt*(dx_1 + dx_2)/2.0
    y[i] = y[i-1] + dt*(dy_1 + dy_2)/2.0

plt.plot(x,y)
plt.show()
```

## Método de Runge-Kutta

```python
n = 2*314
dt = 0.01

x = np.zeros(n)
y = np.zeros(n)

x[0] = x_0
y[0] = y_0

for i in range(1,n):
    dx_1, dy_1 = f(x[i-1], y[i-1])
    dx_2, dy_2 = f(x[i-1] + dt*dx_1/2.0, y[i-1]+dt*dy_1/2.0)
    dx_3, dy_3 = f(x[i-1] + dt*dx_2/2.0, y[i-1]+dt*dy_2/2.0)
    dx_4, dy_4 = f(x[i-1] + dt*dx_3, y[i-1]+dt*dy_3)
    
    x[i] = x[i-1] + dt*(dx_1 + 2*dx_2 + 2*dx_3 + dx_4)/6.0
    y[i] = y[i-1] + dt*(dy_1 + 2*dy_2 + 2*dy_3 + dy_4)/6.0

plt.plot(x,y)
plt.show()
```

## Sistema de Lorenz 

Queremos resolver o seguinte sistema:

$$ \begin{cases} 
    \dot{x} &= \sigma (y - x)\\
    \dot{y} &= x(\rho - z) - y\\
    \dot{z} &= xy - \beta z
\end{cases}$$

onde $\sigma = 10$, $\beta = \dfrac{8}{3}$ e $\rho = 28$. As condições iniciais são $$ (x_0, y_0, z_0) = (1,1,1) $$

```python
rho = 28.0
sigma = 10.0
beta = 8.0 / 3.0

x_0 = 1
y_0 = 1
z_0 = 1

def f(x, y, z):
    dx = rho*(y - x)
    dy = x*(rho - z) - y
    dz = x*y - beta*z
    
    return dx, dy, dz
```

```python
n = 10000
dt = 0.01

x = np.zeros(n)
y = np.zeros(n)
z = np.zeros(n)

x[0] = x_0
y[0] = y_0
z[0] = z_0

for i in range(1, n):
    dx_1, dy_1, dz_1 = f(x[i-1], y[i-1], z[i-1])
    dx_2, dy_2, dz_2 = f(x[i-1] + dt*dx_1/2.0, y[i-1]+dt*dy_1/2.0, z[i-1]+dt*dz_1/2.0)
    dx_3, dy_3, dz_3 = f(x[i-1] + dt*dx_2/2.0, y[i-1]+dt*dy_2/2.0, z[i-1]+dt*dz_2/2.0)
    dx_4, dy_4, dz_4 = f(x[i-1] + dt*dx_3, y[i-1]+dt*dy_3, z[i-1]+dt*dz_3)
    
    x[i] = x[i-1] + dt*(dx_1 + 2*dx_2 + 2*dx_3 + dx_4)/6.0
    y[i] = y[i-1] + dt*(dy_1 + 2*dy_2 + 2*dy_3 + dy_4)/6.0
    z[i] = z[i-1] + dt*(dz_1 + 2*dz_2 + 2*dz_3 + dz_4)/6.0

ax = plt.figure().add_subplot(projection='3d')
ax.plot(x, y, z, 'r', label='Lorenz')
ax.legend()
plt.show()
```

## Usando scipy

```python
rho = 28.0
sigma = 10.0
beta = 8.0 / 3.0

def f(state, t):
    x, y, z = state  # Unpack the state vector
    return sigma * (y - x), x * (rho - z) - y, x * y - beta * z  # Derivatives

state0 = [1.0, 1.0, 1.0]
t = np.arange(0.0, 40.0, 0.01)

states = odeint(f, state0, t)

fig = plt.figure()
ax = fig.gca(projection="3d")
ax.plot(states[:, 0], states[:, 1], states[:, 2])
plt.draw()
plt.show()
```

## Van der Pol oscillator

Queremos resolver $$ \ddot{x} - \mu \dot{x} (1 - x^2)  + x = 0$$ onde $\mu>0$ é uma constante.

Note que: $$ \ddot{x} + \mu \dot{x} (x^2 - 1) = \frac{d}{dt}\left( \dot{x} + \mu \left(\frac{1}{3}x^3 - x\right) \right)$$

Fazendo $w = \dot{x} + \mu F(x)$, onde $F(x) = \dfrac{1}{3}x^3 - x$, logo $\dot{w} = -x$ e temos o sistema
$$ \begin{cases}
    \dot{x} = w - \mu F(x) \\
    \dot{w} = -x
\end{cases}$$

Para resolver vamos usar:

Fazendo $y = \dot{x}$, vamos obter o seguinte sistema equivalente
$$ \begin{cases}
    \dot{x} = y \\
    \dot{y} = \mu y (1-x^2)- x
\end{cases}$$

```python
mu = 1

def f(state, t):
    x,y = state
    dx = y
    dy = mu*y*(1-x*x) - x
    return dx, dy

init = [0.1, 0.1]

t = np.arange(0.0, 20.0, 0.01)
sols = odeint(f, init, t)
plt.plot(sols[:, 0],sols[:, 1])
plt.show()
```

## Atrator de Rössler

Temos o seguinte sistema:
$$ \begin{cases} 
    \dot{x} = - y - z\\
    \dot{y} = x + a y\\
    \dot{z} = b + z(x - c)\\
\end{cases} $$ onde $a = 0.1$, $b = 0.1$ e $c = 14$. 

```python
a = 0.2
b = 0.2
c = 5.7

def f(state, t):
    x,y,z = state
    
    dx = - y - z
    dy = x + a*y
    dz = b + z*(x - c)
    
    return dx, dy, dz

init = [1, 1, 1]
t = np.arange(0.0, 150.0, 0.01)
sols = odeint(f, init, t)

fig = plt.figure()
ax = fig.gca(projection="3d")
ax.plot(sols[:, 0], sols[:, 1], sols[:, 2])
plt.draw()
plt.show()
```
